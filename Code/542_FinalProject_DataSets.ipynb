```python
from sklearn.utils import shuffle
df_true_shuffle = shuffle(df_true_init)
df_fake_shuffle = shuffle(df_fake_init)
```

## Add labels to the data to detect true/fake - T/F

```python
df_true_shuffle.drop('Label', axis=1, inplace=True)
df_true_shuffle.insert(2,'Label','T')
```

```python
df_true_shuffle
```

```python
df_fake_shuffle.drop('Label', axis=1, inplace=True)
df_fake_shuffle.insert(2,'Label','F')
```

```python
df_fake_shuffle
```

##Split 10% of the data to prepare test and train sets and merge the subsequent data

```python
df_fake_test = df_fake_shuffle.iloc[:int(df_fake_shuffle.shape[0]/10):]
df_fake_train = df_fake_shuffle.iloc[int(df_fake_shuffle.shape[0]/10):,:]
```

```python
df_true_test = df_true_shuffle.iloc[:int(df_true_shuffle.shape[0]/10):]
df_true_train = df_true_shuffle.iloc[int(df_true_shuffle.shape[0]/10):,:]
```

```python
df_TestSet = pd.concat([df_fake_test, df_true_test], axis=0)
df_TrainSet = pd.concat([df_fake_train, df_true_train], axis=0)
```

```python
df_TestSet = shuffle(df_TestSet)
df_TrainSet = shuffle(df_TrainSet)
```

```python
print(df_TrainSet['Label'].value_counts(),df_TestSet['Label'].value_counts(),sep='\n\n')
```
