```python
import pandas as pd
df_TrainSet = pd.read_csv(filepath+'/Train.csv') 
df_TestSet = pd.read_csv(filepath+'/Test.csv') 
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
import pandas
import spacy
nlp = spacy.load("en_core_web_sm")

df_TrainSet['proceText'] = (df_TrainSet["text"][:2]
 .apply(lambda text: " "
        .join(token.lemma_ for token in nlp(text) if
              not token.is_punct
              and not token.is_currency
              and not token.is_digit
              and not token.is_punct
              and not token.is_oov
              and not token.is_space
              and not token.is_stop
              and not token.like_num
              and not token.pos_ == "PROPN")))
```

## Preprocess of train data

```python
import spacy
nlp = spacy.load('en_core_web_sm')

def preprocessing(content): 
  #print(content)
  process_list = []
  for line in content:
    #lemmatize, remove stop words and punctuation, remove proper nouns
    filtered = [token.lemma_ for token in nlp(line) if token.is_stop == False and token.text.isalpha() == True and token.pos_ != 'PROPN']
    #filtered = [token.text for token in nlp(content) if token.is_stop == False and token.text.isalpha() == True]
    process_list.append(filtered)
  return process_list
```

```python
train_text_processed_lst = preprocessing(df_TrainSet["text"])
```

```python
df_TrainSet['Text_proc'] = train_text_processed_lst
train_title_processed_lst = preprocessing(df_TrainSet["title"])
df_TrainSet['Title_proc'] = train_title_processed_lst
df_TrainSet.to_csv(filepath+'/TrainProcessed.csv')
```

## Word cloud generation for data analysis

```python
df_TrainSet_proc['Text_proc'][0]
```

```python
#Code to generate word cloud
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud
def generatewordcloud(sentence_lst):
  #print(sentence_lst)
  token_list = sum(sentence_lst, [])#merge list into single list of words
  sentence = " ".join(token_list) #merge list of word into a sentence for wordcloud
  word_cloud = WordCloud(width = 1500, height = 700, collocations = False, background_color = 'black').generate(sentence)
  plt.figure(figsize = (15, 7), facecolor = None)
  plt.imshow(word_cloud)
  plt.axis("off")
  plt.tight_layout(pad = 0)
  plt.show()
```

```python
generatewordcloud(df_TrainSet_proc['Text_proc'])
```

```python
generatewordcloud(df_TrainSet_proc['Text_proc']+df_TrainSet_proc['Title_proc'])
```
